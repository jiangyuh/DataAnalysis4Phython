```python
print("1.")
from string import *
def countWords(s):
    w=s.split()
    return (len(w)) 

filename=open("ques2.txt",'r') #open an file in reading mode
total=0
for line in filename:
    total=total + countWords(line)
print(total)
filename.close()
```

```python
print("2.")

def mode(l):
        count_dict={};
        for i in l:
            if i in count_dict:
                count_dict[i]+=1;
            else:
                count_dict[i]=1;
        max_appear=0
        for v in count_dict.values():
            if v>max_appear:
                max_appear=v;
        if max_appear==1:
            return;
        mode_list=[];
        for k,v in count_dict.items():
            if v==max_appear:
                mode_list.append(k);
        return mode_list;

filename=open("ques2.txt",'r')
l=filename.readline()
wl=l.split()
wordlist=[]
for x in wl:
    wordlist.append(len(x))   

count=0
for y in wordlist:
    count=count+y
print("The average is:")
print(count/len(wordlist))
print("The median is:")
wordlist=sorted(wordlist)
print(wordlist[len(wordlist)//2])
print("the mode is")
print(mode(wordlist))
filename.close()
```

```python
print("1.")
import os
import json
import re
files = os.listdir("jsons")
d={}
for item in files:
    item=item.replace('.json','')
    for txt in open('jsons/' + item +'.json','r') :
        txt= re.sub(r"(,?)(\w+?)\s*?:", r"\1'\2':", txt)
        txt= txt.replace("'", "\"")
        d[item]=json.loads(txt)
subd={}
for item in files:
    first=0
    third=0
    dd={"first":0,"third":0}
    item=item.replace('.json','')
    if "retargetingSegments" in d[item].keys():
        for y in d[item]["retargetingSegments"]:
            for (key, value) in y.items():
                if key=="segmentUid" and value.startswith("0"):
                    first=first+1
                if key=="segmentUid" and value.startswith("1"):
                    third=third+1
    dd["first"]=first
    dd["third"]=third
    subd[item]=dd
print(subd)        
```
